```python
#read authors and conferences csv
author_1 = pd.read_csv("articles_authors_1.csv")
author_2 = pd.read_csv("articles_authors_2.csv")
author_3 = pd.read_csv("articles_authors_3.csv")
confer = pd.read_csv("conferences.csv")
```

```python
# contac author1, author2, author3
authors = pd.concat([author_1,author_2, author_3], ignore_index=True)
print(authors.shape)

# check the number of null values of authors
print(authors.isna().sum())
```

```python
#Take a look at the confer table
print(confer.shape)
```

```python
# merge authors and conferences
merged_df = pd.merge(authors, confer, how = 'inner', left_on=['article_id', 'title'], right_on = ['article_id', 'title'])
print(merged_df.shape)
print(merged_df.isna().sum())
```
